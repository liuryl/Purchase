# 可视化变量关系
regplot方便观察变量间的关系，默认散点图
lmplot x和y必须以字符串的形式传入

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
iris=sns.load_dataset('iris')
```

```python
titannic=sns.load_dataset('titanic')
```

```python
tips
```

```python
iris
```

```python
titannic
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,order=3)
#回归曲线显示出两个数据间的关系
```

# 单变量分布图像，组合图像

```python
sns.distplot(a=titannic.age,rug=True)
#ked曲线可以关闭rug能够看到分布位置
```

```python
sns.distplot(a=titannic.age,rug=True)
```

```python
sns.histplot(data=titannic,x='age',hue='who')
```

```python
sns.kdeplot(data=titannic,x='age')
```

```python
sns.rugplot(data=titannic,x='age')
```

# 双变量分布

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

# 成对变量分布
pairplot

```python
iris
```

```python
sns.pairplot(data=iris)
```

# 多层面版

```python
tips
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='smoker')
```

```python
#构造网格布局
factgrid=sns.FacetGrid(data=tips,row='day',col='size',hue='smoker')
factgrid.map(plt.scatter,'total_bill','tip').add_legend()
```

# 字画布的分割  横轴按照sex  纵轴按照time
# 每个画布绘制散点图，绘制散点图表示total_bill tip的关系并按照smoker分组

```python
factgrid=sns.FacetGrid(data=tips,row='sex',col='time',hue='smoker')

factgrid.map(plt.scatter,'total_bill','tip').add_legend()
```
