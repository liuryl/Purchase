```python
import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
data=pd.read_csv('/Users/liuriyilang/Desktop/课件/机器学习/机器学习和算法/01 KNN机器学习入门/代码/cancer.csv',sep='\t')
data.head()
```

```python
y=data['Diagnosis']
X=data.iloc[:,2:]
display(X.head(),y.head())
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
display(X_train.shape,X_test.shape)
```

# 网格搜索超参数

```python
%%time

estimator=KNeighborsClassifier()

params=dict(n_neighbors=np.arange(1,30),
            weights=['uniform','distance'],
            p=[1,2])

gCV=GridSearchCV(estimator,params,cv=6,scoring='accuracy')
gCV.fit(X_train,y_train)
```

```python
gCV.best_score_
```

```python
gCV.best_params_
```

```python
best_model=gCV.best_estimator_
best_model
```

### 模型预测

```python
y_pre=best_model.predict(X_test)
print('算法预测',y_pre[:20])
```

```python
print('the true value',y_test[:20].values)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pre)
```

```python
(y_test==y_pre).mean()
```

```python
best_model.score(X_test,y_test)
```

```python
gCV.score(X_test,y_test)
```

## 归一化

##### 方法一

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()
X_norml=mms.fit_transform(X)
X_norml
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_norml,y,test_size=0.2,random_state=100)
```

```python
estimator=KNeighborsClassifier()
params=dict(n_neighbors= np.arange(1,30),weights= ['uniform','distance'],p=[1,2])
gCV=GridSearchCV(estimator,params,cv=5,scoring='accuracy')
gCV.fit(X_train,y_train)#训练
```

```python
print('超参数',gCV.best_params_)
```

```python
score=gCV.score(X_test,y_test)
print('min-max归一化数据，模型得分',score)
```

##### 方式二

```python
from sklearn.preprocessing import StandardScaler
mms=StandardScaler()
X_norml2=mms.fit_transform(X)
X_norml2
X_train,X_test,y_train,y_test=train_test_split(X_norml2,y,test_size=0.2,random_state=100)

estimator=KNeighborsClassifier()

params=dict( n_neighbors=np.arange(1,30),
weights=['uniform','ditance'],p=[1,2])

gCV=GridSearchCV(estimator,params,cv=5,scoring='accuracy')
gCV.fit(X_train,y_train)#训练

print('超参数',gCV.best_params_)

score=gCV.score(X_test,y_test)
print('z-score标准化归一化数据模型得分',score)
```
